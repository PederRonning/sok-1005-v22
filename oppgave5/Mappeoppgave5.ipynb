```python
from bs4 import BeautifulSoup
import requests
import numpy as np

def fetch_html_tables(url):
    "Returns a list of tables in the html of url"
    page = requests.get(url)
    bs=BeautifulSoup(page.content)
    tables=bs.find_all('table')
    return tables

tables=fetch_html_tables('https://www.tomshardware.com/best-picks/best-mining-gpus-benchmarked-and-ranked')
table_html=tables[0]

print(str(table_html)[:1000])
```

Laster inn nødvendige pakker og definerer en funksjon vi bruker for å laste inn dataen vi skal bruke.

Henter dataen jeg har tenkt å bruke fra tomshardware.com

```python
def html_to_table(html):
    "Returns the table defined in html as a list"
    #defining the table:
    table=[]
    #iterating over all rows
    for row in html.find_all('tr'):
        r=[]
        #finding all cells in each row:
        cells=row.find_all('td')
        
        #if no cells are found, look for headings
        if len(cells)==0:
            cells=row.find_all('th')
            
        #iterate over cells:
        for cell in cells:
            cell=format(cell)
            r.append(cell)
        
        #append the row to t:
        table.append(r)
    return table

def format(cell):
    "Returns a string after converting bs4 object cell to clean text"
    if cell.content is None:
        s=cell.text
    elif len(cell.content)==0:
        return ''
    else:
        s=' '.join([str(c) for c in cell.content])
        
    #here you can add additional characters/strings you want to 
    #remove, change punctuations or format the string in other
    #ways:
    s=s.replace('\xa0','')
    s=s.replace('\n','')
    return s

table=html_to_table(table_html)

#printing top
print(str(table)[:1000])
```

Skiller kolonnene i tabellen

```python
def save_data(file_name,table):
    "Saves table to file_name"
    f=open(file_name,'w')
    for row in table:
        f.write(';'.join(row)+'\n')
    f.close()
    
save_data('GPU.csv',table)
```

```python
data['Avg. eBay Price (Jan 2022)'] = data['Avg. eBay Price (Jan 2022)'].str.replace('$', '')
data['Avg. eBay Price (Jan 2022)'] = data['Avg. eBay Price (Jan 2022)'].str.replace(',', '')
data['Avg. eBay Price (Jan 2022)']= data['Avg. eBay Price (Jan 2022)'].astype(float)
data['Mining Performance']= data['Mining Performance'].astype(float)
```

Fjerner $ tegn og komma for å gjøre kolonnene numeriske (float)

```python
from matplotlib import pyplot as plt

fig,ax=plt.subplots()

ax.set_ylabel('Mining Performance')
ax.set_xlabel('GPU Price Ebay $')

ax.scatter(data['Avg. eBay Price (Jan 2022)'], data['Mining Performance'],  label='GPU')
ax.legend(loc='lower right',frameon=False)
```

```python
from matplotlib import pyplot as plt

fig,ax=plt.subplots()

ax.set_ylabel('Mining Performance')
ax.set_xlabel('GPU Price Ebay $ (Log)')

ax.scatter(np.log(data['Avg. eBay Price (Jan 2022)']), data['Mining Performance'],  label='GPU')
ax.legend(loc='lower right',frameon=False)
```

```python
A=data['Mining Performance']
pd.DataFrame(A)
```

```python
B=pd.DataFrame(np.log(data['Avg. eBay Price (Jan 2022)']))
B['intercept']=1
B
```

Lager en regresjonsmoddel og legger til verdiene A og B som er Mining performance og gjennomsnittlig pris på ebay.

```python
from statsmodels.regression.linear_model import OLS

res=OLS(A,B).fit()

print(res.summary())
```

```python
B=np.linspace(min(np.log(data['Avg. eBay Price (Jan 2022)'])), max(np.log(data['Avg. eBay Price (Jan 2022)'])), 100)

regression_line=res.params['intercept']+res.params['Avg. eBay Price (Jan 2022)']*B

ax.plot(B, regression_line,color='red')
fig
```
